```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.read_csv("FASTDataFrameNice.csv")
df.head()
```

```
fig, axarr = plt.subplots(5, 5, sharex=True, sharey=True, figsize=(14,14))
for conf_qual in range(1,6):
  for i in range(1,6):
    dfi = df.loc[df.actual_qual==i, ]
    ax = axarr[i-1, conf_qual-1]
    ax.hist(dfi['Quality '+str(conf_qual)]);
    ax.set_title("%i / %i" % (i, conf_qual))

plt.xlim([0,1])
plt.ylim([0,10])
fig.show()
```

```
fig, axarr = plt.subplots(5, sharex=True, sharey=True, figsize=(14,14))
for i in range(1,6):
 
  dfi = df.loc[df.actual_qual==i, ]
  ax = axarr[i-1]
  ax.hist(dfi['predicted_qual']);
  plottitle = "Quality " + str(i) + " Predictions"
  ax.set_title(plottitle)
  ax.set_ylabel('Num of Imgs')
  plt.setp(ax.get_xticklabels(), fontsize = 14)
  
  

# plt.xlim([0,1])
# plt.ylim([0,10])
plt.tight_layout(pad=1)
fig.show()
```
